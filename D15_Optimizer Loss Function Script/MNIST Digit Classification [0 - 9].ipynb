# 2. Import Dataset

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data(path="mnist.npz")
```

# 4. Data Preparation

```python
X_train.std(),X_test.std()
```

```python
X_train = X_train/255
X_test  = X_test/255
```

```python
X_train.std(),X_test.std()
```

# 5. Model Building

### Step 1: Build the architecture

```python
from keras.models import Sequential
from keras.layers import Flatten,Dense
```

```python
model = Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(units = 50,activation = 'relu'))
model.add(Dense(units = 10,activation = 'softmax'))
model.summary()
```

### Step 2: Model Compilation

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
```

# 6. Model Training

```python
model_training = model.fit(x=X_train,y=y_train,batch_size=None, epochs=10,validation_data=(X_test,y_test))
```

# 7. Model Testing

```python
plt.imshow(X_test[20],cmap ='gray')
plt.show()
```

```python
y_pred = model.predict(X_test)
```

```python
import numpy as np
np.argmax(y_pred[20])
```

# 8. Model Evaluation

```python
model.evaluate(x=X_test, y=y_test)
```

# 9. Model Performance

#### Model Loss Pattern

```python
plt.plot(model_training.history['loss'],label='Train loss')
plt.plot(model_training.history['val_loss'],label='Test loss')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Model Loss Performance')
plt.legend()

plt.show()
```

#### Model Accuracy Performance

```python
plt.plot(model_training.history['accuracy'],label='Train Accuarcy')
plt.plot(model_training.history['val_accuracy'],label='Test Accuracy')

plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Performance')
plt.legend()
plt.show()
```
