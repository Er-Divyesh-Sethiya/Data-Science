## 2.3 Manipulate Strings with Methods

### ========================================================================================

## TERMINOLGY ALERT

**Methods/Functions** - both are similar terms.

##  When to go with Methods?
**To perform some operation on top of my variable.**

*NOTE : Methods always ends with ().*

### ========================================================================================

```python
learner_1 = 'Sanjay'
learner_2 = '    aman'
learner_3 = 'sAdHaam'
learner_4 = '     saran kumar   '
```

```python
len(learner_1),len(learner_2),len(learner_3),len(learner_4)
```

```python
learner_1 = learner_1.upper()
learner_2 = learner_2.lstrip()
learner_3 = learner_3.lower()
learner_4 = learner_4.strip()
```

```python
print(learner_1)
print(learner_2)
print(learner_3)
print(learner_4)
```

```python
len(learner_1),len(learner_2),len(learner_3),len(learner_4)
```
