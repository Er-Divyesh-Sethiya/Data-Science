# 1. Import Necessary libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
```

# 2. Import Data

```python
shampoo_sales = pd.read_csv('sales.csv',index_col=0,parse_dates=True)#,squeeze=True)
shampoo_sales
```

# 3. Data Understanding

```python
shampoo_sales.shape
```

```python
shampoo_sales.dtypes
```

```python
shampoo_sales.isna().sum()
```

```python
plt.figure(figsize=(13,5))
shampoo_sales.plot()
plt.title('Month-wise Shampoo Sales',size = 15)
```

```python
shampoo_sales
```

```python
day_wise_shampoo_sales = shampoo_sales.resample('D').mean()
day_wise_shampoo_sales
```

```python
day_wise_shampoo_sales = day_wise_shampoo_sales.interpolate(method = 'linear')
day_wise_shampoo_sales
```

```python
plt.figure(figsize = (16,5))
plt.plot(day_wise_shampoo_sales['Sales'])
plt.show()
```

## Downsampling - Quarterly

```python
quarterwise_shampoo_sales = day_wise_shampoo_sales.resample('Q').mean()
quarterwise_shampoo_sales
```

```python
plt.figure(figsize = (16,5))
plt.plot(quarterwise_shampoo_sales['Sales'])
plt.show()
```
