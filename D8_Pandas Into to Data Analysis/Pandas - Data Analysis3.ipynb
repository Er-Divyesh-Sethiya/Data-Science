# Chapter 8 - Grouping Vs Pivot Table Vs CrossTab

## ========================================================================

## CLUE to use GroupBy function

### Discrete Vs Continous Datatype:

**Discrete** - Countable but not measurable(No units), cannot be seggregated.

**Continous** - Measarable(Units), can be seggregated.

## ========================================================================

```python
insurance_details = pd.read_csv('insurance.csv')
insurance_details
```

### 8.1 On an average, which region is getting more insurance charges?

```python
insurance_details.groupby(by='region')['charges'].mean().sort_values(ascending = False).round()
```

### 8.2 Which gender is getting more insurance charges on an average?

```python
insurance_details.groupby(by = 'sex')['charges'].mean().round().sort_values(ascending = False)
```

### 8.3 Based on region and smokers, what is the average insurance charges?

```python
insurance_details.groupby(by = ['smoker','region'])['charges'].mean().round()
```

```python
insurance_details.groupby(by = ['smoker','region','children'])['charges'].mean().round()
```

```python
pd.pivot_table(data = insurance_details,values='charges',index='region',columns=['smoker','children'], aggfunc='mean').round()
```

```python
pd.pivot_table(data = insurance_details,values='charges',index='region',columns=['smoker'], aggfunc='mean').round().sort_values(by = 'yes',ascending = False)
```
